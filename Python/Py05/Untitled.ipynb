```python
import random #라이브러리 먼저 가져오기 

class Account: #설계도 시작
    account_count = 0 #클래스 변수 
    
    def __init__(self, name, money): 
        self.name = name
        self.money = money
        self.bank = "SC은행"
        #계좌번호 랜덤 생성 (3-2-6)
        num1 = random.randint(100, 999)
        num2 = random.randint(10, 99)
        num3 = random.randint(100000, 999999)
        self.account_num = f"{num1}-{num2}-{num3}"

        Account.account_count += 1

    def get_account_num(self):  #메서드 
        print(Account.account_count)

    def deposit(self, amount):
        if amount < 1:
            print("입금은 1원 이상만 가능합니다.")

    def withdraw(self, amount):
        if self.money < amount:
            print("잔액이 부족합니다.")
            return
        self.money -= amount

    def display_info(self):
        print(f"은행이름: {self.bank}, 예금주: {self.name}, 계좌번호: {self.account_num}, 잔고: {self.money}원")

          #Q7부터 힘듦***
    
        
        
            
```
