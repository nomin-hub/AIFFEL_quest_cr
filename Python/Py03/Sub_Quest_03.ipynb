```python
import math

while True:
    # 1️⃣ 첫 번째 숫자 입력
    while True:
        try:
            num1 = int(input("첫 번째 정수를 입력하세요: "))
            break
        except ValueError:
            print("잘못된 입력입니다. 정수를 입력해주세요.")

    # 2️⃣ 두 번째 숫자 입력
    while True:
        try:
            num2 = int(input("두 번째 정수를 입력하세요: "))
            break
        except ValueError:
            print("잘못된 입력입니다. 정수를 입력해주세요.")

    # 3️⃣ 연산자 입력
    op = input("연산자를 입력하세요 (+, -, *, /, **): ")

    # 4️⃣ 계산
    try:
        if op == "+":
            result = num1 + num2

        elif op == "-":
            result = num1 - num2

        elif op == "*":
            result = num1 * num2

        elif op == "/":
            if num2 == 0:
                raise ZeroDivisionError
            result = num1 / num2

        elif op == "**":
            result = math.pow(num1, num2)

        else:
            print("지원하지 않는 연산자입니다.")
            continue

        print(f"결과: {result}")

    except ZeroDivisionError:
        print("0으로 나눌 수 없습니다.")

    # 5️⃣ 계속할지 묻기
    again = input("계속 계산할까요? (y/n): ")
    if again.lower() != "y":
        print("프로그램 종료")
        break
```
